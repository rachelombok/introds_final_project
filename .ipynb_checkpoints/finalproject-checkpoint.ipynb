## Data Processing  
cleaning up data for only 2015, creating a new dataframe from that

```python
from sklearn.linear_model import LinearRegression

# df_2015 = pd.DataFrame(data=(df.loc[df['Year']==2015]))

df.fillna(0, inplace=True)
```

## Data Modelling
Here is our linear regression model between (variable 1) and (variable 2).
y = b0 + b1*x

- hypothesis testing on features selected

```python
linear_reg = LinearRegression()
x = df['Schooling'].values.reshape(-1, 1)
y = df['Life expectancy '].values.reshape(-1, 1)

linear_reg.fit(x, y)

b0 = linear_reg.predict(([[1]]))
print("b0: ", b0)

b1 = linear_reg.coef_
print("b1: ", b1)

x_array = np.arange(min(df['Schooling']), max(df['Schooling'])).reshape(-1, 1)
plt.scatter(x, y)
y_head = linear_reg.predict(x_array)
plt.plot(x_array, y_head, color="red")
plt.show()
```
